```python
import pandas as pd
import numpy as np
import statistics

from skimage.io import imread
from skimage.io import imshow
from skimage.feature import hog
from skimage.color import rgb2gray
from skimage.transform import resize
from skimage import exposure

import cv2 as cv
import matplotlib.pyplot as plt

from google.colab import drive
from google.colab import files
```

```python
drive.mount('/content/drive')
PATH = "/content/drive/My Drive/candidates" #@param{type:"string"}

from glob import glob
images = [plt.imread(file, cv.IMREAD_GRAYSCALE) for file in glob(PATH + '/*.png')]
print(len(images))
```

```python
plt.figure(figsize = (12, 40), dpi=300)
tick_lable_size = 12

for i in range(len(images)):
  plt.subplot(6, 2, i+1)
 
  
  plt.rc('xtick',labelsize=tick_lable_size)
  plt.rc('ytick',labelsize=tick_lable_size)
  
  images[i] = resize(images[i], (256, 256))
  plt.imshow(images[i])
```
